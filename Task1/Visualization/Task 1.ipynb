```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing Channel ID Dataset

```python
url_Task1_6 = "https://raw.githubusercontent.com/rizwankareem006/youtube_trending_analysis/main/Task1/Task1_6/output/part-r-00000"
Task1_6 = pd.read_csv(url_Task1_6, sep = "\t", names=["Channel_ID", "Channel_Name"])
```

# Task 1.1: Find the channel which got the highest number of videos in the trending list

## Importing Dataset 

```python
url_Task_1_1 = "https://raw.githubusercontent.com/rizwankareem006/youtube_trending_analysis/main/Task1/Task1_1/output/part-r-00000"
Task1_1 = pd.read_csv(url_Task_1_1, sep = "\t", names=["Channel_ID", "No_of_Appearances"])

merged_Task1_1 = pd.merge(Task1_6, Task1_1, on = 'Channel_ID').sort_values("No_of_Appearances", ascending = False)
merged_Task1_1
```

## Creating Figure and Axes

```python
fig_1, (ax1_1, ax1_2) =   plt.subplots(nrows=1, ncols=2, figsize=(20,10))
no_of_appearances = merged_Task1_1.No_of_Appearances.to_numpy()
ax1_1.hist(no_of_appearances,histtype = 'bar', facecolor = 'c', edgecolor = 'k', alpha = 0.8)
ax1_1.set_xlabel('Number of Appearances', fontsize = 15)
ax1_1.set_ylabel('Number of Channels', fontsize = 15)
ax1_2.boxplot(no_of_appearances)
ax1_2.set_xlabel("No of Appearances", fontsize = 15)
ax1_2.set_ylabel("No of Channels", fontsize = 15)
```

## Finding the Channel with Highest Number of Appearances

```python
print("The Channel with Highest Number of Appearances is")
max_apperances = merged_Task1_1.head(1)
max_apperances
```

## Top 20 Channels on the Trending List

```python
x = merged_Task1_1['Channel_Name'].head(20)
y = merged_Task1_1['No_of_Appearances'].head(20)
fig, ax = plt.subplots(figsize = (10,8))
ax.barh(x, y, 0.8, facecolor = 'c', edgecolor = 'k')
plt.gca().invert_yaxis()
ax.grid(b = True, color ='grey', linestyle ='--', linewidth = 0.6, alpha = 0.2)
for i, v in enumerate(y):
    ax.text(v + 3, i + .25, str(v), color = 'k')
ax.set_title('Top 20 Channels on the Trending List', fontsize = 15, fontweight ='bold')
plt.ylabel('Channel Title', fontsize = 15)
plt.xlabel('Number of Appearances', fontsize = 15)
plt.show()
```
